```python
import urllib
from bs4 import BeautifulSoup

import re

import time

start_time = time.time()

url_pattern = "https://www.rfet.es/es/clubes/#PROVINCIA#"
    
provincias = ["alava", "Albacete", "Alicante", "Almeria", "Asturias", "avila", "Badajoz", "Barcelona", "Burgos", "Caceres", "Cadiz", "Cantabria", "Castellon", "Ciudad-Real", "Cordoba", "Cuenca", "Girona", "Granada", "Guadalajara", "guipuzcoa", "Huelva", "Huesca", "Baleares", "Jaen", "a-Coruna", "La-Rioja", "Las-Palmas", "Leon", "Lleida", "Lugo", "Madrid", "Malaga", "Murcia", "Navarra", "Orense", "Palencia", "Pontevedra", "Salamanca", "sc-de-tenerife", "Segovia", "Sevilla", "Soria", "Tarragona", "Teruel", "Toledo", "Valencia", "Valladolid", "Vizcaya", "Zamora", "Zaragoza"];

club_dict = dict()

for provincia in provincias:
    
    print(provincia)
    
    query_url = url_pattern.replace('#PROVINCIA#', provincia.lower())
    
    getpage = urllib.request.urlopen(query_url)
    
    web_page = getpage.read()

    soup = BeautifulSoup(web_page, 'html.parser')
    
    club_list = soup.findAll('div', {'class':'generic-list'})
    
    club_list = str(club_list[0])
    
    club_list = re.compile('<a href=\"([^\"]*)\" title=\"([^\"]*)\"><span>[^<]*<\/span><span>([^<]*)<\/span><\/a>', re.IGNORECASE | re.MULTILINE).findall(club_list)
    
    for club in club_list:
        club_dict[club[0]] = {"Nombre" : club[1].lower().capitalize(), "Provincia" : provincia.capitalize(),"Localidad" : club[2].lower().capitalize()}
    
    
url_pattern = "https://www.rfet.es#CLUB#"

for club in club_dict.keys():
    
    query_url = url_pattern.replace('#CLUB#', club.lower())
    
    print(query_url)
    
    getpage = urllib.request.urlopen(query_url)
    
    web_page = getpage.read()

    soup = BeautifulSoup(web_page, 'html.parser')
    
    club_list = soup.findAll('div', {'class':'club-detail-profile-data'})
    
    str_to_replace = ['\t', '\n', '<p>', '</p>', '<divclass=\"club-detail-profile-data\">', '</div>']
    
    club_list = str(club_list[0])
    
    for char in str_to_replace:
        club_list = club_list.replace(char, '')
        
    club_list = club_list.replace('</a>', '<br/>')
    club_list = club_list.replace('<b>', '<br/>')
    club_list = club_list.replace('</b>', '<br/>')
       
    club_list = club_list.split('<br/>')
    
    club_list = list(filter(None, club_list))
       
    for i in range(len(club_list)):
        if club_list[i] == 'Direccin:':
            if i + 2 < len(club_list) and i + 1 < len(club_list):
                club_dict[club]['Direccion'] = club_list[i + 1] + ' - ' + club_list[i + 2]
        
        if club_list[i] == 'Contacto:':
            if i + 1 < len(club_list):
                if 'Tel.' in club_list[i + 1]:
                    club_dict[club]['Tlf'] = club_list[i + 1].split(' ')[1]
        
        if '<a href="mailto:' in club_list[i]:
            club_dict[club]['Email'] = club_list[i].split('>')[1]        
        
        if '<a href="https:' in club_list[i]:
            club_dict[club]['Web-Site'] = club_list[i].split('>')[1]
            
        if 'Pistas:' in club_list[i]:
            if i + 1 < len(club_list):
                club_dict[club]['Pistas'] = club_list[i + 1]   
            
            
print(club_dict)

f = open("./Output/Club-List-CSV.csv", "w")
f.write("Enlace, Nombre, Provincia, Localidad, Direccion, Tlf, Email, Web-Site, Pistas\r")

for club in club_dict:
    
    URL_RFET = url_pattern.replace('#CLUB#', club.lower())
    
    if 'Nombre' in club_dict[club].keys():
        Nombre = club_dict[club]['Nombre']
    else:
        Nombre = '-'
         
    if 'Provincia' in club_dict[club].keys():
        Provincia = club_dict[club]['Provincia']
    else:
        Provincia = '-'
        
    if 'Localidad' in club_dict[club].keys():
        Localidad = club_dict[club]['Localidad']
    else:
        Localidad = '-'
    
    if 'Direccion' in club_dict[club].keys():
        Direccion = club_dict[club]['Direccion'].replace(',', '')
    else:
        Direccion = '-'
        
    if 'Tlf' in club_dict[club].keys():
        Tlf = club_dict[club]['Tlf']
    else:
        Tlf = '-'

    if 'Email' in club_dict[club].keys():
        Email = club_dict[club]['Email']
    else:
        Email = '-'
        
    if 'Web-Site' in club_dict[club].keys():
        Web_Site = club_dict[club]['Web-Site']
    else:
        Web_Site = '-'
        
    if 'Pistas' in club_dict[club].keys():
        Pistas = club_dict[club]['Pistas']
    else:
        Pistas = '-'
    
    #    
    csv_line = (URL_RFET  + ", " +
                Nombre    + ", " +
                Provincia + ", " +
                Localidad + ", " +
                Direccion + ", " +
                Tlf       + ", " +
                Email     + ", " +
                Web_Site  + ", " +
                Pistas    + "\r")
    
    f.write(csv_line)
    
f.close()

print("Finished!")

print('Processing time: ' + str(time.time()-start_time) + ' s')
```
